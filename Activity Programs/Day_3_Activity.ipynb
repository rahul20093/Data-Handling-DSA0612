1) Weather Data

---



```r
weather_data <- data.frame(Date = c("2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"),
                           Temperature = c(10, 12, 8, 15, 14),
                           Humidity = c(75, 70, 80, 65, 72),
                           Wind_Speed = c(15, 12, 18, 20, 16))

library(ggplot2)
ggplot(weather_data, aes(x = Date, y = Temperature, fill = Humidity)) +
  geom_tile() +
  scale_fill_gradient(low = "blue", high = "red")

library(ggplot2)
ggplot(weather_data, aes(x = Date, y = Temperature)) +
  geom_violin()

library(ggplot2)
ggplot(weather_data, aes(x = Date, y = Temperature)) +
  geom_boxplot()
```

2)  Financial Market Data

---



```r
financial_data <- data.frame(Date = c("2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"),
                             Stock_Price = c(100, 102, 98, 105, 108),
                             Volume_Traded = c(2.5, 3, 2.2, 2.8, 3.5),
                             Market_Cap = c(500, 510, 490, 525, 540))

mean(financial_data$Stock_Price)

mean(financial_data$Stock_Price)

library(ggplot2)
ggplot(financial_data, aes(x = Date, y = Stock_Price)) +
  geom_bar(stat = "identity")
```

3) Consumer Survey Data

---



```r
library(ggplot2)

data <- data.frame(
  Product = c("A", "B", "C", "D", "E"),
  Price = c(50, 70, 60, 45, 55),
  Rating = c(4.2, 3.8, 4.0, 4.5, 3.9)
)

ecdf_plot <- ggplot(data, aes(Price)) +
  stat_ecdf(geom = "step") +
  labs(title = "ECDF of Product Prices", x = "Price", y = "ECDF")

print(ecdf_plot)

density_hist <- ggplot(data, aes(Price)) +
  geom_histogram(aes(y = ..density..), bins = 5, fill = "lightblue", color = "black") +
  geom_density(alpha = 0.2, fill = "red") +
  labs(title = "Density Histogram of Product Prices", x = "Price", y = "Density")

print(density_hist)

skewed_plot <- ggplot(data, aes(Rating)) +
  geom_histogram(bins = 5, fill = "lightgreen", color = "black") +
  labs(title = "Distribution of Product Ratings", x = "Rating", y = "Count")

print(skewed_plot)
```
