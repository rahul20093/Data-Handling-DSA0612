6)Create a scatter plot where each point represents a customer, with demographics on one axis and purchase behavior on the other. Use different colors or shapes to represent different clusters identified by the algorithm. Interpretation involves analyzing the clusters to understand distinct customer segments, such as high-spending young adults versus budget-conscious seniors.



---



```r
set.seed(123)
library(ggplot2)
library(cluster)
data <- data.frame(Age = rnorm(100, 35, 10), Spend = rnorm(100, 500, 150))
clusters <- hclust(dist(data), method = "ward.D2")
data$Cluster <- cutree(clusters, k = 3)
ggplot(data, aes(x = Age, y = Spend, color = as.factor(Cluster))) + geom_point()
```

7)Create a time series plot of historical sales data with a forecasted line extending into the future. Add confidence intervals to indicate the range of uncertainty in the forecast. Interpretation involves assessing the forecast accuracy and adjusting predictions based on changing trends or external factors.


---



```r
set.seed(123)
library(ggplot2)
sales <- cumsum(rnorm(24, mean = 100, sd = 10))
ts_data <- ts(sales, frequency = 12)
forecast <- HoltWinters(ts_data)
plot(forecast)
```

8)Create a bar chart where each bar represents a branch and the height of the bar represents the KPI value. Use grouped or stacked bars to compare multiple metrics across branches. Interpretation involves identifying top-performing branches and potential areas for improvement based on performance metrics.



---



```r
library(ggplot2)
branches <- data.frame(Branch = c("A", "B", "C"), Revenue = c(50000, 60000, 55000))
ggplot(branches, aes(x = Branch, y = Revenue)) + geom_bar(stat = "identity", fill = "skyblue")
```

9)Create a scatter plot where each point represents a department, with training hours on the x-axis and productivity metrics on the y-axis. Add a trend line or regression line to visualize the overall relationship. Interpretation involves assessing whether departments with higher training hours also demonstrate higher productivity based on the slope and direction of the trend line.



---



```r
library(ggplot2)
training <- data.frame(Hours = runif(10, 10, 40), Productivity = runif(10, 50, 100))
ggplot(training, aes(x = Hours, y = Productivity)) + geom_point() + geom_smooth(method = "lm")
```
