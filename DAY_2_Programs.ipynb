11) Area Plot

---



```r
library(ggplot2)

revenue_data <- data.frame(
  Year = 2015:2019,
  Revenue = c(30, 35, 40, 45, 50)
)

ggplot(revenue_data, aes(x = Year, y = Revenue)) +
  geom_area(fill = "skyblue", alpha = 0.5) +
  labs(title = "Yearly Revenue", x = "Year", y = "Revenue (in million)") +
  theme_minimal()
```

12) Step Plot

---



```r
library(ggplot2)

cumulative_sales_data <- data.frame(
  Month = factor(c("Jan", "Feb", "Mar", "Apr", "May"), levels = c("Jan", "Feb", "Mar", "Apr", "May")),
  Cumulative_Sales = c(100, 200, 300, 400, 500)
)

ggplot(cumulative_sales_data, aes(x = Month, y = Cumulative_Sales, group = 1)) +
  geom_step() +
  labs(title = "Cumulative Sales Over Months", x = "Month", y = "Cumulative Sales") +
  theme_minimal()
```

13) Ridgeline Plot

---



```r
library(ggridges)

temperature_data <- data.frame(
  City = rep(c("City1", "City2"), each = 3),
  Temperature = c(20, 21, 19, 22, 23, 24)
)

ggplot(temperature_data, aes(x = Temperature, y = City, fill = City)) +
  geom_density_ridges() +
  labs(title = "Temperature Distributions", x = "Temperature", y = "City") +
  theme_minimal()
```

14) Dumbbell Plot

---



```r
library(ggalt)

sales_campaign_data <- data.frame(
  Product = c("A", "B", "C", "D", "E"),
  Before = c(200, 300, 400, 500, 600),
  After = c(250, 350, 450, 550, 650)
)

ggplot(sales_campaign_data) +
  geom_dumbbell(aes(x = Before, xend = After, y = Product),
                size = 3, color = "gray", size_x = 3, size_xend = 3) +
  labs(title = "Sales Before and After Campaign", x = "Sales", y = "Product") +
  theme_minimal()
```

15) Lollipop Plot

---



```r
library(ggplot2)

sales_region_data <- data.frame(
  Region = c("North", "South", "East", "West"),
  Sales = c(150, 200, 180, 210)
)

ggplot(sales_region_data, aes(x = Region, y = Sales)) +
  geom_segment(aes(x = Region, xend = Region, y = 0, yend = Sales), color = "gray") +
  geom_point(color = "blue", size = 5) +
  labs(title = "Sales by Region", x = "Region", y = "Sales") +
  theme_minimal()
```

16) Spaghetti Plot

---



```r
library(ggplot2)

sales_time_data <- data.frame(
  Month = rep(c("Jan", "Feb", "Mar"), 2),
  Product = rep(c("A", "B"), each = 3),
  Sales = c(100, 150, 130, 120, 140, 160)
)

ggplot(sales_time_data, aes(x = Month, y = Sales, group = Product, color = Product)) +
  geom_line(size = 1) +
  geom_point(size = 3) +
  labs(title = "Sales over Time by Product", x = "Month", y = "Sales") +
  theme_minimal()
```

17) Waterfall Plot

---



```r
library(ggplot2)

profit_loss_data <- data.frame(
  labels = c("Jan", "Feb", "Mar", "Apr", "May", "Total"),
  values = c(100, -20, 50, -10, 80, 0)
)

profit_loss_data$end <- cumsum(profit_loss_data$values)
profit_loss_data$start <- c(0, head(profit_loss_data$end, -1))
profit_loss_data$type <- ifelse(profit_loss_data$values >= 0, "positive", "negative")
profit_loss_data$type[length(profit_loss_data$type)] <- "total"

ggplot(profit_loss_data, aes(x = labels, y = values, fill = type)) +
  geom_rect(aes(x = labels, xmin = labels, xmax = labels,
                ymin = pmin(start, end), ymax = pmax(start, end)),
            colour = "black", size = 0.25) +
  geom_hline(yintercept = 0) +
  scale_fill_manual(values = c("positive" = "green", "negative" = "red", "total" = "blue")) +
  labs(title = "Profit and Loss Waterfall", x = "", y = "Amount") +
  theme_minimal() +
  theme(legend.position = "none")
```

20) Calendar Heatmap

---



```r
library(ggplot2)
library(lubridate)

activity_data <- data.frame(
  Date = as.Date(c("2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05")),
  Count = c(10, 12, 15, 8, 20)
)

activity_data$Weekday <- wday(activity_data$Date, label = TRUE)
activity_data$Week <- week(activity_data$Date)

ggplot(activity_data, aes(x = Weekday, y = Week, fill = Count)) +
  geom_tile(color = "white") +
  scale_fill_gradient(low = "white", high = "blue") +
  labs(title = "Daily Activity Count", x = "Weekday", y = "Week", fill = "Count") +
  theme_minimal()
```
