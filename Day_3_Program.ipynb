1) Weather data

---



```r
library(ggplot2)
library(reshape2)

data <- data.frame(
  Date = as.Date(c("2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05")),
  Temperature = c(10, 12, 8, 15, 14),
  Humidity = c(75, 70, 80, 65, 72),
  WindSpeed = c(15, 12, 18, 20, 16)
)

ggplot(data, aes(x = Humidity, y = WindSpeed, z = Temperature)) +
  geom_tile(aes(fill = Temperature)) +
  geom_contour(aes(z = Temperature), color = "white") +
  theme_minimal() +
  labs(title = "Temperature vs Humidity and Wind Speed", fill = "Temp (°C)")
```

2) Financial Market Data

---



```r
library(ggplot2)

financial_data <- data.frame(
  Date = as.Date(c("2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05")),
  Stock_Price = c(100, 102, 98, 105, 108),
  Volume_Traded = c(2.5, 3, 2.2, 2.8, 3.5),
  Market_Cap = c(500, 510, 490, 525, 540)
)

ggplot(financial_data, aes(x = Stock_Price, y = Volume_Traded, z = Market_Cap)) +
  geom_point() +
  theme_classic() +
  labs(title = "Financial Market Data", x = "Stock Price ($)", y = "Volume Traded (millions)", z = "Market Capitalization ($)")
```

3) Consumer Survey Data

---



```r
library(ggplot2)

consumer_data <- data.frame(
  Product = factor(c("A", "B", "C", "D", "E")),
  Price = c(50, 70, 60, 45, 55),
  Rating = c(4.2, 3.8, 4, 4.5, 3.9),
  Age_Group = factor(c("25-35", "35-45", "18-25", "45-55", "25-35"))
)

ggplot(consumer_data, aes(x = Price, y = Rating, color = Age_Group)) +
  geom_point() +
  theme_classic() +
  labs(title = "Consumer Survey Data", x = "Price ($)", y = "Rating (out of 5)", color = "Age Group")
```

4) Environmental Data

---



```r
library(ggplot2)

environmental_data <- data.frame(
  Location = factor(c("A", "B", "C", "D", "E")),
  Temperature = c(15, 20, 18, 12, 17),
  Humidity = c(65, 70, 68, 60, 72),
  CO2_Levels = c(400, 450, 420, 380, 430)
)

ggplot(environmental_data, aes(x = Temperature, y = Humidity, z = CO2_Levels)) +
  geom_point() +
  theme_classic() +
  labs(title = "Environmental Data", x = "Temperature (°C)", y = "Humidity (%)", z = "CO2 Levels (ppm)")
```

5) Academic Performance Data

---



```r
library(ggplot2)

academic_data <- data.frame(
  Student = factor(c("A", "B", "C", "D", "E")),
  Math_Score = c(85, 72, 90, 78, 88),
  Science_Score = c(78, 85, 80, 75, 82),
  Attendance = c(95, 92, 98, 85, 93)
)

ggplot(academic_data, aes(x = Math_Score, y = Science_Score, color = Attendance)) +
  geom_point() +
  theme_classic() +
  labs(title = "Academic Performance Data", x = "Math Score", y = "Science Score", color = "Attendance (%)")
```
